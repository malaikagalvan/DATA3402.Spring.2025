```python


class counter:
    def __init__(self,value,max_value):
        self.value=value
        self.max_value=max_value

    def increment(self):
        if self.value>=self.max_value:
            print("Error the value is too large")
        else:
            self.value+=1
            
    def reset(self):
        self.value=0
    
    def get_value(self):
        return self.value
    
```

```python
## Canvas Class
import math
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self):
        self.shapes=list()
        
    def add_shape(self,shape):
        self.shapes.append(shape)
    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
    def paint(self,canvas):
        for shape in self.shapes:
            shape.paint(canvas)
        
            
```

```python
#Testing
c4=Canvas(50,40)
rd=RasterDrawing()

rect=Rectangle(9,8,9,10,char="*")
cir=Circle(5,10,12,char="*")
tri=Triangle(5,3,2,3,4,5,char="*")

rd.add_shape(rect)
print(rd.shapes)
rd.add_shape(cir)
print(rd.shapes)
rd.add_shape(tri)
print(rd.shapes)


rd.paint(c4)
c4.display()

#modify
rect=Rectangle(20,10,8,6,char="^")
rd.shapes[0]=rect
rd.update(c4)
c4.display()
```

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```

```python
## Shapes


import math
class Shapes:
    def __init__(self,x,y, **kwargs):
        self.__x=x
        self.__y=y
        self.kwargs=kwargs
    
    def __repr__(self):
        return f'shapes({repr(self.x())},{repr(self.y())},{repr(self.kwargs)})'

    def paint(self,canvas): pass
    #points on perimeter 
    def points_perimeter(self):
        raise NotImplementedError
    #inside object
    def inside_object(self):
        raise NotImplementedError   
    def area(self):
        raise NotImplementedError
    def perimeter(self):
        raise NotImplementedError

    def get_x(self):
        return self.__x
    def get_y(self):
        return self.__y

class Rectangle(Shapes):
    def __init__(self,length,width,x,y,**kwargs):
        super().__init__(x,y,**kwargs)
        self.__length=length
        self.__width=width
    
    def __repr__(self):
        return f'Rectangle({repr(self.x())}, {repr(self.y())}, {repr(self.__length)}, {repr(self.__width)}, {repr(self.kwargs)})'
    
    def length(self):
        return self.__length
    def width(self):
        return self.__width
    
    
    def area(self):
        return self.__length*self.__width
    def perimeter(self):
        rect_perimeter=2*self.__length+2*self.__width
        return rect_perimeter
    
    def inside_object(self,x,y):
        return self.x()<=x<=self.x()+self.__length and self.y()<=y<=self.y()+self.__width

    
    #Points on perimeter Rectangle
    def points_perimeter(self):
        list_points=list()
        
        bottom_left=(self.x(), self.y())
        bottom_right=(self.__length+self.x(),self.y())
        top_right=(self.__length+self.x(), self.y()+self.__width)
        top_left=(self.x(),self.y()+self.__width)

        
        for i in range(4):
            x_coord=bottom_left[0]+i*(self.__length/4)
            y_coord=bottom_left[1]
            list_points.append((x_coord,y_coord))
        for i in range(4):
            x_coord= bottom_right[0]
            y_coord=bottom_right[1]+i*(self.__width/4)
            list_points.append((x_coord,y_coord))
        for i in range(4):
            x_coord=top_right[0]-i*(self.__length/4)
            y_coord=top_right[1]
            list_points.append((x_coord,y_coord))
        for i in range(4):
            x_coord=top_left[0]
            y_coord=top_left[1]-i*(self.__width/4)
            list_points.append((x_coord,y_coord))
    
        return list_points
    
    
    def get_length(self):
        return self.__length
    def get_width(self):
        return self.__width
    
    def paint(self, canvas):
        canvas.v_line(self.__length, self.__width,self.x(),**self.kwargs)
        canvas.v_line(self.__length, self.__width+self.y(),self.x(),**self.kwargs)
        canvas.h_line(self.__length, self.__width,self.y(),**self.kwargs)
        canvas.h_line(self.__length+self.x(),self.__width, self.y(),**self.kwargs)
        
class Circle(Shapes):
    def __init__(self,radius,x,y, **kwargs):
        super().__init__(x,y,**kwargs)
        self.__radius=radius
    
    def __repr__(self):
        return f'Circle({repr(self.x())}, {repr(self.y())}, {repr(self.__radius)}, {repr(self.kwargs)})'

    def area(self):
        return 3.14*self.__radius**2
    def perimeter(self):
        return 2*3.14*self.__radius
    
    def radius(self):
        return self.__radius
    
    # Points for Circle
    def points_perimeter(self):
        list_points=list()
        num_points=16
        for i in range(num_points):
            angle=2*math.pi*i/num_points
            x_coord=self.x()+self.__radius*math.cos(angle)
            y_coord=self.y()+self.__radius*math.sin(angle)
            list_points.append((round(x_coord), round(y_coord)))
        return list_points
    
    def inside_object(self,x,y):
        
        return (x-self.x())**2+(y-self.y())**2<=self.__radius**2
    
    def paint(self,canvas):
        points=self.points_perimeter()
        for x_coord, y_coord in points:
             if 0 <= x_coord < canvas.width and 0 <= y_coord < canvas.height:
                canvas.set_pixel(y_coord, x_coord)
    def get_radius(self):
        return self.__radius 
        
class Triangle(Shapes):
    def __init__(self,a,b,c,h,x,y,**kwargs):
        super().__init__(x,y,**kwargs)
        self.__a=a
        self.__b=b
        self.__c=c
        self.__h=h
     
    def __repr__(self):
        return f'Triangle({repr(self.x())}, {repr(self.y())}, {repr(self.__a)}, {repr(self.__b)}, {repr(self.__c)}, {repr(self.__h)}, {repr(self.kwargs)})'

    
    
    def area(self):
        return (self.__b*self.__h)/2
    def perimeter(self):
        return self.__a+self.__b+self.__c
    
    # Points
    def points_perimeter(self):
        list_points=list()
        v1=self.x(),self.y()
        v2=self.x()+self.__b,self.y()
        v3=self.x()+self.__b/2,self.y()+self.__h
        
        for i in range(5):
            x_coord=v1[0]+i*(self.__b/5)
            y_coord=v1[1]
            list_points.append((round(x_coord), round(y_coord)))
        for i in range(5):
            x_coord=v2[0]-i*(self.__b/5)
            y_coord=v2[1]+i*(self.__h/5)
            list_points.append((round(x_coord), round(y_coord)))
        for i in range(6):
            x_coord=v3[0]-i*(self.__b/6)
            y_coord=v3[1]-i*(self.__h/6)
            list_points.append((round(x_coord), round(y_coord)))
        return list_points
    
   
    
    #Paint Triangle
    def paint(self,canvas):
        points=self.points_perimeter()
        
        for x_coord, y_coord in points:
            if 0 <= x_coord < canvas.width and 0 <= y_coord < canvas.height:
                canvas.set_pixel(y_coord, x_coord)
    
class CompoundShape(Shapes):
    def __init__(self,shape):
        self.shape=shape
    def paint(self,canvas):
        for s in self.shape:
            s.paint(canvas)
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes=list()
        
    def add_shape(self,shape):
        self.shapes.append(shape)
    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
    def paint(self,canvas):
        for shape in self.shapes:
            shape.paint(canvas)
    def save(self,filename):
         with open(filename, 'w') as f:
            for shape in self.shapes:
                f.write(repr(shape)+'\n')
    def load(self,filename):
         with open(filename, 'r') as f:
            lines = f.readlines()
            self.shapes = [eval(line.strip()) for line in lines]
```
